# ELM

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib notebook

import seaborn as sns
```

```python
result_file_pre = './elm_multi_output/elm_preprocessed_result.csv'
result_file_raw = './elm_multi_output/elm_raw_result.csv'

f_pre = pd.read_csv(result_file_pre)
f_raw = pd.read_csv(result_file_raw)

result_pre = f_pre['RMSE'].tolist()
result_raw = f_raw['RMSE'].tolist()
```

# Preprocessed

```python
plot_pre = sns.distplot(result_pre, norm_hist=True, kde=True, rug=True)
plt.title('ELM - preprocessed', fontsize=13)
plt.xlabel('RMSE score', fontsize=10)
plt.ylabel('Density', fontsize=10)
```

```python
plt.close()
```

```python
mean_pre = np.mean(result_pre)
std_pre = np.std(result_pre)

print 'number of test samples = ',
print str(len(result_pre))
print

print 'mean value of TestAcc with preprocessed data = ',
print str(mean_pre)
print 'std value of TestAcc with preprocessed data = ',
print str(std_pre)
print

print 'min Test RMSE = ',
print str(min(result_pre))
print 'max Test RMSE = ',
print str(max(result_pre))
```

# Raw

```python
plot_raw = sns.distplot(result_raw, norm_hist=True, kde=True, rug=True, color='g')
plt.title('ELM - raw', fontsize=13)
plt.xlabel('RMSE score', fontsize=10)
plt.ylabel('Density', fontsize=10)
```

```python
plt.close()
```

```python
mean_raw = np.mean(result_raw)
std_raw = np.std(result_raw)

print 'number of test samples = ',
print str(len(result_raw))
print

print 'mean value of TestAcc with raw data = ',
print str(mean_raw)
print 'std value of TestAcc with raw data = ',
print str(std_raw)
print

print 'min Test RMSE = ',
print str(min(result_raw))
print 'max Test RMSE = ',
print str(max(result_raw))
```

# Total Result

```python
plot_pre = sns.distplot(result_pre, norm_hist=True, kde_kws={"label":"preprocessed"})
plot_raw = sns.distplot(result_raw, norm_hist=True, kde_kws={"label":"raw"})
plt.title('ELM', fontsize=13)
plt.xlabel('RMSE score', fontsize=10)
plt.ylabel('Density', fontsize=10)

ax = plt.xlim(0, 1)
```
