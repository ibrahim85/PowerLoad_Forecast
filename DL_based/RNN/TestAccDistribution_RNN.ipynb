# Recurrent Neural Network

```python
import cPickle as pickle
import numpy as np

import matplotlib.pyplot as plt
%matplotlib notebook
import seaborn as sns
```

```python
result_file_pre = 'rnn_preprocessed_result.bin'
result_file_raw = 'rnn_raw_result.bin'

f_pre = open(result_file_pre)
f_raw = open(result_file_raw)

result_pre = pickle.load(f_pre)
result_raw = pickle.load(f_raw)
```

# Preprocessed

```python
plot_pre = sns.distplot(result_pre, hist=True, kde=True, rug=True)
plt.title('Recurrent Neural Network - preprocessed', fontsize=13)
plt.xlabel('RMSE score', fontsize=10)
plt.ylabel('Prob', fontsize=10)
```

```python
plt.close()
```

```python
mean_pre = np.mean(result_pre)
std_pre = np.std(result_pre)

print 'number of test samples = ',
print str(len(result_pre))
print

print 'mean value of TestAcc with preprocessed data = ',
print str(mean_pre)
print 'std value of TestAcc with preprocessed data = ',
print str(std_pre)
print

print 'min Test RMSE = ',
print str(min(result_pre))
print 'max Test RMSE = ',
print str(max(result_pre))
```

# Raw

```python
plot_raw = sns.distplot(result_raw, hist_kws={'color':'g'}, kde_kws={'color':'g'}, rug_kws={'color':'g'})
plt.title('Recurrent Neural Network - raw', fontsize=13)
plt.xlabel('RMSE score', fontsize=10)
plt.ylabel('Prob', fontsize=10)
```

```python
plt.close()
```

```python
mean_raw = np.mean(result_raw)
std_raw = np.std(result_raw)

print 'number of test samples = ',
print str(len(result_raw))
print

print 'mean value of TestAcc with raw data = ',
print str(mean_raw)
print 'std value of TestAcc with raw data = ',
print str(std_raw)
print

print 'min Test RMSE = ',
print str(min(result_raw))
print 'max Test RMSE = ',
print str(max(result_raw))
```

# Total Result

```python
plot_pre = sns.distplot(result_pre, hist=True, kde_kws={"label":"preprocessed"}, rug=True)
plot_raw = sns.distplot(result_raw, hist=True, kde_kws={"label":"raw"}, rug=True)
plt.title('Recurrent Neural Network', fontsize=13)
plt.xlabel('RMSE score', fontsize=10)
plt.ylabel('Prob', fontsize=10)

ax = plt.xlim(0, 0.2)
```

```python
plt.close()
```
