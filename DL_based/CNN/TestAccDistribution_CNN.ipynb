# Convolutional Neural Network

```python
import cPickle as pickle
import numpy as np

import matplotlib.pyplot as plt
%matplotlib notebook
import seaborn as sns
```

```python
# file directories
result_file_pre_00 = 'cnn_preprocessed_result_00.bin'
result_file_pre_01 = 'cnn_preprocessed_result_01.bin'
result_file_pre_02 = 'cnn_preprocessed_result_02.bin'

result_file_raw_00 = 'cnn_raw_result_00.bin'
result_file_raw_01 = 'cnn_raw_result_01.bin'
result_file_raw_02 = 'cnn_raw_result_02.bin'

# open files
f_pre_00 = open(result_file_pre_00)
f_pre_01 = open(result_file_pre_01)
f_pre_02 = open(result_file_pre_02)

f_raw_00 = open(result_file_raw_00)
f_raw_01 = open(result_file_raw_01)
f_raw_02 = open(result_file_raw_02)

# load data
result_pre_00 = pickle.load(f_pre_00)
result_pre_01 = pickle.load(f_pre_01)
result_pre_02 = pickle.load(f_pre_02)

result_raw_00 = pickle.load(f_raw_00)
result_raw_01 = pickle.load(f_raw_01)
result_raw_02 = pickle.load(f_raw_02)

# concatonate data
result_pre = np.hstack((result_pre_00, result_pre_01))
result_pre = np.hstack((result_pre, result_pre_02))

result_raw = np.hstack((result_raw_00, result_raw_01))
result_raw = np.hstack((result_raw, result_raw_02))
```

# Preprocessed

```python
plot_pre = sns.distplot(result_pre, norm_hist=True, kde=True, rug=True)
plt.title('Convolutional Neural Network - preprocessed', fontsize=13)
plt.xlabel('RMSE score', fontsize=10)
plt.ylabel('Density', fontsize=10)
```

```python
plt.close()
```

```python
mean_pre = np.mean(result_pre)
std_pre = np.std(result_pre)

print 'number of test samples = ',
print str(len(result_pre))
print

print 'mean value of TestAcc with preprocessed data = ',
print str(mean_pre)
print 'std value of TestAcc with preprocessed data = ',
print str(std_pre)
print

print 'min Test RMSE = ',
print str(min(result_pre))
print 'max Test RMSE = ',
print str(max(result_pre))
```

# Raw

```python
plot_raw = sns.distplot(result_raw, norm_hist=True, kde=True, rug=True, color='g')
plt.title('Convolutional Neural Network - raw', fontsize=13)
plt.xlabel('RMSE score', fontsize=10)
plt.ylabel('Density', fontsize=10)
```

```python
plt.close()
```

```python
mean_raw = np.mean(result_raw)
std_raw = np.std(result_raw)

print 'number of test samples = ',
print str(len(result_raw))
print

print 'mean value of TestAcc with raw data = ',
print str(mean_raw)
print 'std value of TestAcc with raw data = ',
print str(std_raw)
print

print 'min Test RMSE = ',
print str(min(result_raw))
print 'max Test RMSE = ',
print str(max(result_raw))
```
