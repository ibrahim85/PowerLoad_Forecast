```python
import cPickle as pickle
import numpy as np

import matplotlib.pyplot as plt
%matplotlib notebook
import seaborn as sns
```

```python
result_file_pre = 'multiple_preprocessed_result.bin'
result_file_raw = 'multiple_raw_result.bin'

f_pre = open(result_file_pre)
f_raw = open(result_file_raw)

result_pre = pickle.load(f_pre)
result_raw = pickle.load(f_raw)
```

# Preprocessed

```python
sns.distplot(result_pre, hist=True, kde=True, rug=True)
```

```python
mean_pre = np.mean(result_pre)
std_pre = np.std(result_pre)

print 'mean value of TestAcc with preprocessed data = ',
print str(mean_pre)
print 'std value of TestAcc with preprocessed data = ',
print str(std_pre)
```

```python
sns.distplot(result_raw, hist=True, kde=True, rug=True)
```

```python
mean_raw = np.mean(result_raw)
std_raw = np.std(result_raw)

print 'mean value of TestAcc with raw data = ',
print str(mean_raw)
print 'std value of TestAcc with raw data = ',
print str(std_raw)
```
